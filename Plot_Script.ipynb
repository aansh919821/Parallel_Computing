```python
import numpy as np
import matplotlib.pyplot as plt
import os

# List of output files
files = [
    'output_64_64_64_3_8_1.txt', 'output_64_64_64_3_8_2.txt', 'output_64_64_64_3_8_3.txt', 'output_64_64_64_3_8_4.txt', 'output_64_64_64_3_8_5.txt',
    'output_64_64_64_3_16_1.txt', 'output_64_64_64_3_16_2.txt', 'output_64_64_64_3_16_3.txt', 'output_64_64_64_3_16_4.txt', 'output_64_64_64_3_16_5.txt',
    'output_64_64_64_3_32_1.txt', 'output_64_64_64_3_32_2.txt', 'output_64_64_64_3_32_3.txt', 'output_64_64_64_3_32_4.txt', 'output_64_64_64_3_32_5.txt',
    'output_64_64_64_3_64_1.txt', 'output_64_64_64_3_64_2.txt', 'output_64_64_64_3_64_3.txt', 'output_64_64_64_3_64_4.txt', 'output_64_64_64_3_64_5.txt',
    'output_64_64_96_7_8_1.txt', 'output_64_64_96_7_8_2.txt', 'output_64_64_96_7_8_3.txt', 'output_64_64_96_7_8_4.txt', 'output_64_64_96_7_8_5.txt',
    'output_64_64_96_7_16_1.txt', 'output_64_64_96_7_16_2.txt', 'output_64_64_96_7_16_3.txt', 'output_64_64_96_7_16_4.txt', 'output_64_64_96_7_16_5.txt',
    'output_64_64_96_7_32_1.txt', 'output_64_64_96_7_32_2.txt', 'output_64_64_96_7_32_3.txt', 'output_64_64_96_7_32_4.txt', 'output_64_64_96_7_32_5.txt',
    'output_64_64_96_7_64_1.txt', 'output_64_64_96_7_64_2.txt', 'output_64_64_96_7_64_3.txt', 'output_64_64_96_7_64_4.txt', 'output_64_64_96_7_64_5.txt',
]

# Function to extract times from each file
def extract_times(file_path):
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None, None, None

    Read_times = []
    Main_Code_times = []
    total_times = []

    for i in range(2, len(lines), 3):
        try:
            iteration_times = list(map(float, lines[i].strip().split(', ')))
            if len(iteration_times) == 3:
                Read_times.append(iteration_times[0])
                Main_Code_times.append(iteration_times[1])
                total_times.append(iteration_times[2])
        except (ValueError, IndexError):
            continue

    if not Read_times or not Main_Code_times or not total_times:
        print(f"No valid time data in {file_path}")
        return None, None, None

    return np.array(Read_times), np.array(Main_Code_times), np.array(total_times)

# Organize the times by nc and np
nc_values = [3, 7]
np_vals = [8, 16, 32, 64]

# Initialize dictionaries to store times
time_data = {
    3: {'Read': {p: [] for p in np_vals},
        'Main_Code': {p: [] for p in np_vals},
        'Total': {p: [] for p in np_vals}},
    7: {'Read': {p: [] for p in np_vals},
        'Main_Code': {p: [] for p in np_vals},
        'Total': {p: [] for p in np_vals}}
}

for file in files:
    try:
        parts = file.split('_')
        nc = int(parts[4])
        p = int(parts[5])
        file_path = os.path.join('/content/drive/MyDrive/output3/', file)

        Read_times, Main_Code_times, total_times = extract_times(file_path)
        if Read_times is None:
            continue

        avg_Read_time = np.mean(Read_times)
        avg_Main_Code_time = np.mean(Main_Code_times)
        avg_total_time = np.mean(total_times)

        time_data[nc]['Read'][p].append(avg_Read_time)
        time_data[nc]['Main_Code'][p].append(avg_Main_Code_time)
        time_data[nc]['Total'][p].append(avg_total_time)
    except Exception as e:
        print(f"Error processing {file}: {str(e)}")
        continue

# Calculate averages
avg_times = {
    nc: {
        category: [np.mean(time_data[nc][category][p]) if time_data[nc][category][p] else 0
                 for p in np_vals]
        for category in ['Read', 'Main_Code', 'Total']
    }
    for nc in nc_values
}

print("Averages for nc=3:")
print(f"Read times: {avg_times[3]['Read']}")
print(f"Main Code times: {avg_times[3]['Main_Code']}")
print(f"Total times: {avg_times[3]['Total']}")

print("\nAverages for nc=7:")
print(f"Read times: {avg_times[7]['Read']}")
print(f"Main Code times: {avg_times[7]['Main_Code']}")
print(f"Total times: {avg_times[7]['Total']}")

```

```python
# Plotting the results with specified colors and styles
# [Previous code remains the same until the plotting section]

# Plotting the results
plt.figure(figsize=(12, 8))

# Plot for nc=3
plt.subplot(2, 1, 1)
plt.errorbar(np_vals, avg_times[3]['Read'], label='Read Time', marker='o', linestyle='-', color='b', capsize=5)
plt.errorbar(np_vals, avg_times[3]['Main_Code'], label='Main_Code Time', marker='s', linestyle='-', color='r', capsize=5)
plt.errorbar(np_vals, avg_times[3]['Total'], label='Total Time', marker='^', linestyle='-', color='g', capsize=5)
plt.title('Performance for nc=3')
plt.xlabel('Number of Processes (p)')
plt.ylabel('Time (seconds)')
plt.xticks(np_vals)
plt.grid(True)
plt.legend()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# [Previous code for file processing remains the same until the time_data dictionary]

# Calculate averages and standard deviations
avg_times = {
    nc: {
        category: [np.mean(time_data[nc][category][p]) if time_data[nc][category][p] else 0
                 for p in np_vals]
        for category in ['Read', 'Main_Code', 'Total']
    }
    for nc in nc_values
}

std_devs = {
    nc: {
        category: [np.std(time_data[nc][category][p]) if time_data[nc][category][p] else 0
                 for p in np_vals]
        for category in ['Read', 'Main_Code', 'Total']
    }
    for nc in nc_values
}

print("Averages and Standard Deviations for nc=3:")
print(f"Read times - Avg: {avg_times[3]['Read']}, Std: {std_devs[3]['Read']}")
print(f"Main Code times - Avg: {avg_times[3]['Main_Code']}, Std: {std_devs[3]['Main_Code']}")
print(f"Total times - Avg: {avg_times[3]['Total']}, Std: {std_devs[3]['Total']}")

print("\nAverages and Standard Deviations for nc=7:")
print(f"Read times - Avg: {avg_times[7]['Read']}, Std: {std_devs[7]['Read']}")
print(f"Main Code times - Avg: {avg_times[7]['Main_Code']}, Std: {std_devs[7]['Main_Code']}")
print(f"Total times - Avg: {avg_times[7]['Total']}, Std: {std_devs[7]['Total']}")

# Plotting the results with error bars showing standard deviation
plt.figure(figsize=(12, 8))

# Plot for nc=3
plt.subplot(2, 1, 1)
plt.errorbar(np_vals, avg_times[3]['Read'], yerr=std_devs[3]['Read'],
             label='Read Time', marker='o', linestyle='-', color='b', capsize=5)
plt.errorbar(np_vals, avg_times[3]['Main_Code'], yerr=std_devs[3]['Main_Code'],
             label='Main_Code Time', marker='s', linestyle='--', color='r', capsize=5)
plt.errorbar(np_vals, avg_times[3]['Total'], yerr=std_devs[3]['Total'],
             label='Total Time', marker='^', linestyle='--', color='g', capsize=5)
plt.title('Performance for nc=3 with Standard Deviation Error Bars')
plt.xlabel('Number of Processes (p)')
plt.ylabel('Time (seconds)')
plt.xticks(np_vals)
plt.grid(True)
plt.legend()

```
